```python
cacen['Full Statement of the Enquiry'].value_counts()
```

```python
cacen = pd.read_excel('exl_data_Y.xlsx',sheetname='Call Centre Data')
```

```python
cacen['id']=0
cacen['Level1']=0
cacen['Level2']=0
cacen['Level3']=0
cacen['Level4']=0
for i in range(0,cacen.shape[0]):
    print(i)
    n=cacen['Full Statement of the Enquiry'][i].index('-')
    cacen['id'][i] = cacen['Full Statement of the Enquiry'][i][0:n]
    cacen['Full Statement of the Enquiry'][i]=cacen['Full Statement of the Enquiry'][i][n+1:]
    n=cacen['Full Statement of the Enquiry'][i].index('-')
    cacen['Level1'][i] = cacen['Full Statement of the Enquiry'][i][0:n]
    cacen['Full Statement of the Enquiry'][i]=cacen['Full Statement of the Enquiry'][i][n+1:]
    n=cacen['Full Statement of the Enquiry'][i].index('-')
    cacen['Level2'][i] = cacen['Full Statement of the Enquiry'][i][0:n]
    cacen['Full Statement of the Enquiry'][i]=cacen['Full Statement of the Enquiry'][i][n+1:]
    n=cacen['Full Statement of the Enquiry'][i].index('-')
    cacen['Level3'][i] = cacen['Full Statement of the Enquiry'][i][0:n]
    cacen['Level4'][i]=cacen['Full Statement of the Enquiry'][i][n+1:]
```
