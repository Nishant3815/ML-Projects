```python
demogy['Equipment Warranty'].fillna('Three Full Years',inplace='True')
```

```python
demogv['Equipment Warranty'].fillna('Three Full Years',inplace='True')
```

```python
demogy['Gender'].fillna('other',inplace=True)
demogv['Gender'].fillna('other',inplace=True)
```

```python
demogy.isnull().sum()
demogy.isnull().sum()
```

```python
demogy['Professional Info'].fillna(0.0,inplace=True)
demogv['Professional Info'].fillna(0.0,inplace=True)
```

```python
demogy = demogy.rename(columns={'Salary Slab':'Salslab','Age of Home':'agehome','Professional Info': 'profinfo'})
demogv = demogv.rename(columns={'Salary Slab':'Salslab','Age of Home':'agehome','Professional Info': 'profinfo'})
```

```python
#Alloting median of each scheme age
demogy.loc[demogy.Age.isnull(), 'Age'] = demogy.groupby('Scheme').Age.transform('median')
demogv.loc[demogv.Age.isnull(), 'Age'] = demogv.groupby('Scheme').Age.transform('median')
```

```python
#Alloting median of each scheme age of home
demogy.loc[demogy.agehome.isnull(), 'agehome'] = demogy.groupby('Scheme').agehome.transform('median')
demogv.loc[demogv.agehome.isnull(), 'agehome'] = demogv.groupby('Scheme').agehome.transform('median')
```

```python
#Alloting median of each scheme age
demogy.loc[demogy.Salslab.isnull(), 'Salslab'] = demogy.groupby('Scheme').Salslab.transform('median')
demogv.loc[demogv.Salslab.isnull(), 'Salslab'] = demogv.groupby('Scheme').Salslab.transform('median')
```
